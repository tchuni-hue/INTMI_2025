```python
numero = int(input("Digite um número inteiro: "))
if numero % 2 == 0:
    print("O número {numero} é par.")
else:
    print("O número {numero} é ímpar.")
```

```python
nome=str(input('Digite o seu nome:'))
if 'a' in nome:
    print('O nome tem a letra a.')
else:
    print('O nome não tem letra a.')
```

```python
numero = int(input("Digite um número: "))
if numero % 5 == 0:
    print("O número {numero} é múltiplo de 5.")
else:
    print("O número {numero} não é múltiplo de 5.")
```

```python
temperatura = float(input("Digite a temperatura atual (°C): "))
if temperatura > 30:
    print("Está quente!")
else:
    print("Está frio!")
```

```python
opcao = (input('Digite "S" para continuar ou "N" para sair: '))
if opcao == "S":
    print("Você escolheu continuar.")
elif opcao == "N":
    print("Você escolheu sair.")
else:
    print("Opção inválida. Digite apenas S ou N.")
```

```python
numero = float(input("Digite um número: "))

if numero > 0:
    print("O número é positivo.")
elif numero == 0:
    print("O número é zero.")
else:
    print("O número é negativo.")
```

```python
numero = int(input("Digite um número de 1 a 7: "))
dias = {
    1: "Domingo",
    2: "Segunda-feira",
    3: "Terça-feira",
    4: "Quarta-feira",
    5: "Quinta-feira",
    6: "Sexta-feira",
    7: "Sábado"}
if numero in dias:
    print("O dia correspondente é {dias[numero]}.")
else:
    print("Número inválido! Digite um número entre 1 e 7.")
```

```python
ano = int(input("Digite um ano: "))
if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print("O ano {ano} é bissexto.")
else:
    print("O ano {ano} não é bissexto.")
```

```python
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_por_hora = float(input("Digite o valor pago por hora: "))
if horas_trabalhadas <= 40:
    salario = horas_trabalhadas * valor_por_hora
else:
    horas_normais = 40
    horas_extras = horas_trabalhadas - horas_normais
    salario = (horas_normais * valor_por_hora) + (horas_extras * valor_por_hora * 1.5)
print(f"O salário total é: R$ {salario:.2f}")
```
