```python
def primo(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
nums = [int(input()) for _ in range(10)]
primos = [n for n in nums if primo(n)]
print(primos)
```

```python
mat = [[int(input()) for _ in range(3)] for _ in range(3)]
identidade = True
for i in range(3):
    for j in range(3):
        if (i == j and mat[i][j] != 1) or (i != j and mat[i][j] != 0):
            identidade = False
print("É identidade?" , identidade)
```

```python
produtos = {"arroz": [5.99, 6.50, 6.20],"feijão": [7.20, 7.00, 7.80],"açúcar": [4.10, 4.50, 4.00]}
maior_var = None
produto_var = ""
for nome, precos in produtos.items():
    variacao = max(precos) - min(precos)
    if maior_var is None or variacao > maior_var:
        maior_var = variacao
        produto_var = nome
print("Maior variação:", produto_var, "(", maior_var, ")")
```
